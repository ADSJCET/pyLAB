Visualising the data in Table

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Species", data=data)

plt.show()
```

Comparing SepalLengthCm & SepalWidthCm of Iris

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data = data)
plt.legend(bbox_to_anchor=(1,1), loc=2)

plt.show()
```

Comparing PetalLengthCm & PetalWidthCm of Iris

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data = data)
plt.legend(bbox_to_anchor=(1,1), loc=2)

plt.show()
```

Multivariate Analysis

```python
sns.pairplot(data.drop(['Id'], axis=1), hue='Species', height=1)
```

### Handing Correlation

it comapres the relationship between two variables 

Using a simple `.corr()` is not suitable for this purpose. We need to use a heatmap to visualize the correlation matrix.

in heatmax we can see the correlation between the variables. The lighter the color, the more correlated the variables are.
look for the lightest color box inside the heatmap mostly near the diagonal line boxes contains [1].

```python
data.corr(method='pearson')
```

```python
sns.heatmap(data.corr(method='pearson').drop(['Id'], axis=1).drop(['Id'], axis=0), annot=True)

plt.show()
```

### whats int64?

int64 is a data type in pandas. It is used to store integer values. It is a 64-bit signed integer. It can store values from -9223372036854775808 to 9223372036854775807.

use `astype` to convert the int64 to int32


```python
import numpy as np

x = np.array([1,2,3], dtype=np.int64)
print("Type of Memory Allocation of x",x.dtype)

y = x.astype(np.int32)
print("Type of Memory Allocation of y",y.dtype)
```
