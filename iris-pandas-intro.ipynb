Visualising the data in Table

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Species", data=data)

plt.show()
```

Comparing SepalLengthCm & SepalWidthCm of Iris

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data = data)
plt.legend(bbox_to_anchor=(1,1), loc=2)

plt.show()
```

Comparing PetalLengthCm & PetalWidthCm of Iris

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data = data)
plt.legend(bbox_to_anchor=(1,1), loc=2)

plt.show()
```
